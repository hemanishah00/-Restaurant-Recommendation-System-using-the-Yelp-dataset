**The General idea behind this notebook originated from the question:
What if the user has not given any ratings to any restraunts? </br>This will always be the case when the user is new to yelp.</br>
We came up with the solution that we will recommend restraunts which are the most popular and the most liked.</br>
We do this by selecting only values which have more than 3000 reviews. We consider only these values to avoid the case where there were very small number of reviews.</br></t> For example: if one user gives a restaurant 5 star rating then it should not be considered to be in our recommendation.</br>
We then sort the values according to the number of stars and then by the number of reviews**

We are finding the restraunts which are most liked and also most reviewed

```python
bus = bus[bus['review_count']>=3000].sort_values(by = "stars", ascending=False)[:10]
```

```python
newuser = itemname.merge(bus, on='business_id', how='inner').sort_values(['stars','review_count'], ascending=False)
```
